```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
x=pd.read_csv("C:/Users/Kumari/OneDrive/Desktop/Healthcare/census_2011.csv")
x
```

```python
x.info()
```

```python
x.isnull()
```

```python
x2=x.fillna(method="pad")
x2
```

```python
x2.isnull()
```

```python
state_population = x2.groupby('State name')['Population'].sum().reset_index()
state_population = state_population.sort_values(by='Population', ascending=False)
num_states = len(state_population)
colors = plt.cm.tab20(np.linspace(0, 1, num_states))
plt.figure(figsize=(12, 12))
plt.pie(state_population['Population'], labels=state_population['State name'], autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Population by State')
plt.legend()

plt.axis('equal')
plt.show()
```

```python
state_population = x2.groupby('State name')['Population'].sum().reset_index()
state_population = state_population.sort_values(by='Population', ascending=False)
num_states = len(state_population)
colors = plt.cm.get_cmap('tab20b', num_states)
plt.figure(figsize=(8,8))
wedges, texts, autotexts = plt.pie(state_population['Population'], labels=state_population['State name'], autopct='%1.1f%%', startangle=140, colors=colors(np.linspace(0, 1, num_states)))
plt.title('Population by State')
plt.legend(wedges, state_population['State name'], title="States", loc="lower left", bbox_to_anchor=(0, 0))
plt.axis('equal')  
plt.show()
```

```python
sns.set()
male_count = x2['Male'].sum()
female_count = x2['Female'].sum()
gender_count = pd.DataFrame({
    'Gender': ['Male', 'Female'],
    'Count': [male_count, female_count]
})
plt.figure(figsize=(8, 8))
colors = sns.color_palette('pastel')[0:2]
plt.pie(gender_count['Count'], labels=gender_count['Gender'], autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Gender Distribution')
total_count = male_count + female_count
plt.text(-1.5, 1.2, f'Total Count: {total_count}', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))
plt.axis('equal')  
plt.show()
```
