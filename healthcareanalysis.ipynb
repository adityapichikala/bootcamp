```python
data = {
    'age0-29': x2['Age_Group_0_29'],
    'age30-49': x2['Age_Group_30_49'],
    'age50+': x2['Age_Group_50']
}

# Convert data to a DataFrame
df = pd.DataFrame(data)

# Reshape the DataFrame for the box plot
df_melted = df.melt(var_name='Age Group', value_name='Values')

# Function to remove outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers from each age group
df_no_outliers = pd.concat([
    remove_outliers(df_melted[df_melted['Age Group'] == 'age0-29'], 'Values'),
    remove_outliers(df_melted[df_melted['Age Group'] == 'age30-49'], 'Values'),
    remove_outliers(df_melted[df_melted['Age Group'] == 'age50+'], 'Values')
])


print(df_no_outliers.head())

plt.figure(figsize=(10, 6))
sns.boxplot(x='Age Group', y='Values', data=df_no_outliers)
plt.title('Box Plot of Different Age Groups (Without Outliers)')
plt.ylabel('Values')
plt.xlabel('Age Group')

plt.show()
```
